# Machine learning implementation 


```python
#Label Encoding 
from sklearn.preprocessing import LabelEncoder
m=LabelEncoder()
data['Species']=m.fit_transform(data['Species'])
```

```python
data
```

```python
data['Species'].value_counts()
```

```python
#Feature selection 
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#divide your data into training and testing 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
```

```python
#create a logistic regression model 
from sklearn.linear_model import LogisticRegression 
model=LogisticRegression()
```

```python
#fitting training data into model 
model.fit(xtrain,ytrain)
```

```python
print(model.predict([[6.1,3.5,0.4,0.2]]))
```

# evaluation of a model

```python
#checking score of training data 
model.score(xtrain,ytrain)*100
```

```python
#checking score of testing data 
model.score(xtest,ytest)*100
```

```python
predictresult=model.predict(xtest)
```

```python
#checking accuracy score of a model 
from sklearn.metrics import accuracy_score 
print(accuracy_score(ytest,predictresult)*100)
```
